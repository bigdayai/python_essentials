# Lesson Tutorial

## <span id="1"></span>1. Overview

Check out [What is Data?](www.google.com) for an introduction to Python Datatypes.

Data is the digital representation of real world phenomena.  There are many different ways of representing real world phenomena: numbers, text, pictures, video, audio, messages, documents, etc.  Each of these are different types of data. Some data types are easier for computers to store, analyze and change while others are more difficult.  This lesson is on Structured Data Types.

### Keywords

* **Structured Data:** Data that follows a consistent format and fits into an existing data model or schema. Structured data can easily fit into spreadsheets.
* **Integer:** Whole numbers
* **Float:** Numbers with a fractional part
* **String:** Combinations of alphabetical and numerical characters in the form of text
* **Boolean:** The logical values of True and False

* **Methods:** Built-in actions that you can use to somehow change data
* **Operators:** Symbols for certain actions that change data: +, -, *, /, etc

## <span id="2"></span>2. Implications & Applications

### Implications

Each type of data allows you to represent the world in a specific way.  Numbers allow you to quantify.  Text allows you to quantify and contextualize. * Regression

### 1: INTEGERS (int)

Integers are whole numbers.  When you check to see the data type of a whole number, the output will be "int," short for integer.

```python
type(3)
```

```python
type(23789401)
```

### Common Arithmetic Operators for Integers

Python provides many different ways to change data.  Here are some example operations you can use with the integer data type.  Making calculations is just one way you can change or "manipulate" data.  Learn more about the integer data type using [Python's documentation.](https://docs.python.org/3/library/stdtypes.html)

| Operation | Meaning |
| :-: | :- |
| + | Addition |
| - | Subtraction |
| * | Multiplication |
| / | Division |
| % | Modulus |
| ** | Power |

**Addition**

```python
3 + 4
```

**Subtraction**

```python
105 - 72
```

**Multiplication**

```python
3 * 45
```

**Division (float)**

```python
35 / 7
```

**Division (floor)**
<br/><br/>
In this example, 35 is not a multiple of 9, so if you divided 35 by 9 the answer would be 3.8889.  However, using the // operator, it automatically rounds down to display only the whole number without the decimal fraction.

```python
35 / 9
```

**Modulus**
<br/><br/>
Returns the remainder of a division equation.  In this example, 4 fits into 5 one time with a remainder of 1.

```python
5 % 4
```

**Power**
<br/><br/>
Multiplies a number by itself a certain number of times.  For example, 3 multiplied by itself 4 times.

```python
3 ** 4
```

### 3: STRINGS (str) 

Strings are combinations of alphabetical and numberic characterics in the form of text.  You can make characters into a string by wrapping them in single or double quote marks.

```python
type("When I was your age, I walked 3 miles in the snow to get to school.")
```

### Common String Methods

The string data type comes with many built-in methods for changing strings.  Here are some of the most common.

**Capitalize** 
<br/><br/>
Use the .capitalize method to capitalize the first letter of a string.

```python
"i am learning python".capitalize()
```

**Casefold** 
<br/><br/>
Use the .casefold method to put all characters into lower case.

```python
"I AM LEARNING PYTHON".casefold()
```

**Count** 
<br/><br/>
Use the .count method to identify the number of times a character or combination of characters appear in a larger string. 

```python
"I AM LEARNING PYTHON".count("A")
```

```python
"I AM LEARNING PYTHON".count("PYTHON")
```

**Endswith**
<br/><br/>
Use the .endswith method to check if a string ends with a specific charcter or combination of characters.

```python
"123 Python Road".endswith("Street")
```

```python
"123 Python Road".endswith("Road")
```

There are many more string methods to help you analyze and manipulate text data.  Learn more by visiting [W3's list of string methods](https://www.w3schools.com/python/python_ref_string.asp) or searching "Python String Methods" in your preferred search engine.

### 4: BOOLEAN (bool)

Boolean values are the backbone of automated decision making.  The two boolean, logical values are True and False.  Many statements can return a boolean value.

```python
type(True)
```

```python
type(False)
```

### Common Comparison Operators for Boolean Logic


| Operation | Meaning |
| :-: | :- |
| < | strictly less than |
| <= | less than or equal |
| > | strictly greater than |
| >= | greater than or equal |
| == | equal |
| != | not equal |
| is | object identity |
| is not | negated object identity |

**Strictly Less Than**

```python
bool(4 < 3)
```

```python
bool(4 < 5)
```

**Less Than or Equal**

```python
bool(4 <= 3)
```

```python
bool(4 <= 4)
```

**Equal**

The equal symbol in Python (==) is a little different from the standard mathematical equal symbol (=) becuase the standard equal symbol (=) actually has a different use which you'll learn about in the next lesson.

```python
bool(15 == 14)
```

```python
bool('A' == 'A')
```

**Not Equal**

Not equal allows you to check the inverse of equal.

```python
bool(15 != 14)
```

```python
bool('A' != 'A')
```

**Is & Is Not**

Is allows you to check the identity of an object.

```python
15 is bool
```

```python
True is bool
```

```python
15 is 15
```

```python
15 is not 15
```
